3.Cardiovascular Health and Neurodegeneration: A study by researchers from Rush University Medical Center and the University of California found that improved cardiovascular health, assessed using the American Heart Association‚Äôs Simple 7 habits (including a healthy diet, regular exercise, non-smoking, and maintaining healthy blood pressure, BMI, and cholesterol levels), may lower the risk of neurodegenerative diseases such as dementia and Alzheimer's.


```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
import optuna
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler, PolynomialFeatures
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from torch.utils.data import DataLoader, TensorDataset
from tabulate import tabulate

# Load dataset
data = pd.read_csv(r"updated_data.csv")

# Encode categorical variables
label_encoders = {}
for col in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])
    label_encoders[col] = le

# Encode target variable
le_target = LabelEncoder()
data['Sleep Disorder'] = le_target.fit_transform(data['Sleep Disorder'])

# Train-test split
X = data.drop(columns=['Sleep Disorder'])
y = data['Sleep Disorder']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Feature Mapping (Polynomial Features)
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_train_mapped = poly.fit_transform(X_train)
X_test_mapped = poly.transform(X_test)

# Standardize numerical features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_mapped)
X_test_scaled = scaler.transform(X_test_mapped)

# Convert data to PyTorch tensors
X_train_tensor = torch.tensor(X_train_scaled, dtype=torch.float32)
X_test_tensor = torch.tensor(X_test_scaled, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train.values, dtype=torch.long)
y_test_tensor = torch.tensor(y_test.values, dtype=torch.long)

class ResNetBlock(nn.Module):
    """Optimized Residual Block with Dropout, Batch Normalization, and L1 Regularization"""
    def __init__(self, input_dim, hidden_dim, dropout_rate=0.3):
        super(ResNetBlock, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.bn1 = nn.BatchNorm1d(hidden_dim)
        self.dropout = nn.Dropout(dropout_rate)
        self.fc2 = nn.Linear(hidden_dim, input_dim)
        self.bn2 = nn.BatchNorm1d(input_dim)
        self.relu = nn.ReLU()

    def forward(self, x):
        identity = x  # Skip connection
        out = self.fc1(x)
        out = self.bn1(out)
        out = self.relu(out)
        out = self.dropout(out)
        out = self.fc2(out)
        out = self.bn2(out)
        out += identity  # Add skip connection
        out = self.relu(out)
        return out

class ResNetOptimizedL1FM(nn.Module):
    """ResNet with Optimizations: Feature Mapping, Dropout, BatchNorm, Hyperparameter Tuning, and L1 Regularization"""
    def __init__(self, input_dim, num_classes, dropout_rate, l1_lambda):
        super(ResNetOptimizedL1FM, self).__init__()
        self.fc1 = nn.Linear(input_dim, 64)
        self.bn1 = nn.BatchNorm1d(64)
        self.relu = nn.ReLU()
        self.resblock1 = ResNetBlock(64, 128, dropout_rate)
        self.resblock2 = ResNetBlock(64, 128, dropout_rate)
        self.fc2 = nn.Linear(64, num_classes)
        self.l1_lambda = l1_lambda  # L1 Regularization Strength
        self.init_weights()  # Apply Xavier Initialization

    def forward(self, x):
        x = self.fc1(x)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.resblock1(x)
        x = self.resblock2(x)
        x = self.fc2(x)
        return x

    def init_weights(self):
        """Apply Xavier initialization to layers"""
        for m in self.modules():
            if isinstance(m, nn.Linear):
                nn.init.xavier_uniform_(m.weight)
                if m.bias is not None:
                    nn.init.zeros_(m.bias)

    def l1_regularization(self):
        """Compute L1 regularization loss"""
        l1_norm = sum(torch.norm(param, 1) for param in self.parameters())
        return self.l1_lambda * l1_norm

def objective(trial):
    """Objective function for hyperparameter tuning"""
    dropout_rate = trial.suggest_float("dropout_rate", 0.1, 0.5)
    learning_rate = trial.suggest_float("learning_rate", 0.0001, 0.01, log=True)
    batch_size = trial.suggest_categorical("batch_size", [16, 32, 64])
    l1_lambda = trial.suggest_float("l1_lambda", 1e-5, 1e-2, log=True)  # L1 Regularization Strength

    # Define model
    model = ResNetOptimizedL1FM(input_dim=X_train_mapped.shape[1], num_classes=len(y.unique()), dropout_rate=dropout_rate, l1_lambda=l1_lambda)

    # Loss, Optimizer, and Learning Rate Scheduler
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)

    # Create DataLoader
    train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

    # Training
    epochs = 20
    for epoch in range(epochs):
        model.train()
        for X_batch, y_batch in train_loader:
            optimizer.zero_grad()
            outputs = model(X_batch)
            loss = criterion(outputs, y_batch) + model.l1_regularization()  # Add L1 loss
            loss.backward()
            optimizer.step()
        scheduler.step()

    # Evaluation
    model.eval()
    with torch.no_grad():
        y_pred = model(X_test_tensor).argmax(dim=1).numpy()
    accuracy = accuracy_score(y_test, y_pred)

    return accuracy

# Run Optuna Study
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=20)
best_params = study.best_params
print("Best Hyperparameters:", best_params)

final_model = ResNetOptimizedL1FM(input_dim=X_train_mapped.shape[1], num_classes=len(y.unique()), dropout_rate=best_params["dropout_rate"], l1_lambda=best_params["l1_lambda"])

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(final_model.parameters(), lr=best_params["learning_rate"])
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.5)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=best_params["batch_size"], shuffle=True)

# Training
epochs = 50
for epoch in range(epochs):
    final_model.train()
    for X_batch, y_batch in train_loader:
        optimizer.zero_grad()
        outputs = final_model(X_batch)
        loss = criterion(outputs, y_batch) + final_model.l1_regularization()  # L1 Loss
        loss.backward()
        optimizer.step()
    scheduler.step()


final_model.eval()

# Training Predictions
with torch.no_grad():
    y_train_pred = final_model(X_train_tensor).argmax(dim=1).numpy()

# Testing Predictions
with torch.no_grad():
    y_test_pred = final_model(X_test_tensor).argmax(dim=1).numpy()


train_accuracy = accuracy_score(y_train, y_train_pred)
train_report = classification_report(y_train, y_train_pred)
train_conf_matrix = confusion_matrix(y_train, y_train_pred)

print("\n=== üöÄ Training Performance ===")
print(f"Training Accuracy: {train_accuracy:.4f}")
print("\nTraining Classification Report:\n", train_report)

# Training Confusion Matrix - Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(train_conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=le_target.classes_, yticklabels=le_target.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Training - ResNet Optimized)')
plt.show()

test_accuracy = accuracy_score(y_test, y_test_pred)
test_report = classification_report(y_test, y_test_pred)
test_conf_matrix = confusion_matrix(y_test, y_test_pred)

print("\n=== üöÄ Testing Performance ===")
print(f"Testing Accuracy: {test_accuracy:.4f}")
print("\nTesting Classification Report:\n", test_report)

# Testing Confusion Matrix - Heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(test_conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=le_target.classes_, yticklabels=le_target.classes_)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Testing - ResNet Optimized)')
plt.show()


def analyze_heart_health_neuro_impact(model, data, poly, scaler, le_target):

    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    import seaborn as sns
    import torch
    import torch.nn.functional as F

    # Define the AHA Simple 7 habits and their effects
    aha_habits = [
        ('Healthy Diet', {'BMI Category': -0.7, 'Systolic': -3, 'Diastolic': -2, 'Quality of Sleep': 0.5}),
        ('Regular Exercise', {'Physical Activity Level': 1.5, 'Daily Steps': 2500, 'Heart Rate': -3, 'Stress Level': -0.8}),
        ('Non-Smoking', {'Heart Rate': -2, 'Systolic': -4, 'Diastolic': -2}),
        ('Healthy Blood Pressure', {'Systolic': -8, 'Diastolic': -5}),
        ('Healthy BMI', {'BMI Category': -1.0, 'Quality of Sleep': 0.4, 'Sleep Duration': 0.3})
    ]

    # Make sure we exclude the target column from our analysis
    feature_cols = [col for col in data.columns if col != 'Sleep Disorder']
    numeric_cols = data[feature_cols].select_dtypes(include=['float64', 'int64']).columns.tolist()

    # Create baseline patient profile (average values)
    baseline_profile = data[numeric_cols].mean().to_dict()

    # Create improved cardiovascular health profile with AHA's Simple 7 adjustments
    improved_profile = baseline_profile.copy()

    # Apply research-based adjustments
    # Physical activity improvements
    if 'Physical Activity Level' in baseline_profile:
        improved_profile['Physical Activity Level'] = min(10, baseline_profile['Physical Activity Level'] + 2.0)
    if 'Daily Steps' in baseline_profile:
        improved_profile['Daily Steps'] = min(15000, baseline_profile['Daily Steps'] + 3000)

    # BMI improvements
    if 'BMI Category' in baseline_profile:
        improved_profile['BMI Category'] = max(1, baseline_profile['BMI Category'] - 1.0)

    # Cardiovascular metrics improvements
    if 'Heart Rate' in baseline_profile:
        improved_profile['Heart Rate'] = max(60, baseline_profile['Heart Rate'] - 5)
    if 'Systolic' in baseline_profile:
        improved_profile['Systolic'] = max(110, baseline_profile['Systolic'] - 10)
    if 'Diastolic' in baseline_profile:
        improved_profile['Diastolic'] = max(70, baseline_profile['Diastolic'] - 6)

    # Stress reduction
    if 'Stress Level' in baseline_profile:
        improved_profile['Stress Level'] = max(1, baseline_profile['Stress Level'] - 1.5)

    # Sleep improvements
    if 'Quality of Sleep' in baseline_profile:
        improved_profile['Quality of Sleep'] = min(10, baseline_profile['Quality of Sleep'] + 1.2)
    if 'Sleep Duration' in baseline_profile:
        improved_profile['Sleep Duration'] = min(8.5, baseline_profile['Sleep Duration'] + 0.7)

    # Convert profiles to DataFrame format
    baseline_df = pd.DataFrame([baseline_profile])
    improved_df = pd.DataFrame([improved_profile])

    # Apply preprocessing pipeline
    try:
        # Apply polynomial features
        baseline_poly = poly.transform(baseline_df)
        improved_poly = poly.transform(improved_df)

        # Apply scaling
        baseline_scaled = scaler.transform(baseline_poly)
        improved_scaled = scaler.transform(improved_poly)

        # Convert to tensors
        baseline_tensor = torch.FloatTensor(baseline_scaled)
        improved_tensor = torch.FloatTensor(improved_scaled)

        # Get predictions
        model.eval()
        with torch.no_grad():
            baseline_logits = model(baseline_tensor)
            improved_logits = model(improved_tensor)

        # Convert logits to probabilities
        baseline_probs = F.softmax(baseline_logits, dim=1).numpy()[0]
        improved_probs = F.softmax(improved_logits, dim=1).numpy()[0]

        # Get class names (keeping them as numbers)
        class_names = np.arange(len(baseline_probs))

        # Create comparison visualization
        plt.figure(figsize=(12, 6))
        x = np.arange(len(class_names))
        width = 0.35

        plt.bar(x - width/2, baseline_probs, width, label='Baseline')
        plt.bar(x + width/2, improved_probs, width, label='Improved Heart Health')

        plt.xlabel('Sleep Disorder Class')
        plt.ylabel('Probability')
        plt.title('Impact of AHA Simple 7 on Sleep Disorder Probabilities')
        plt.xticks(x, class_names)
        plt.legend()
        plt.tight_layout()
        plt.show()
        # Create a table for baseline vs improved probabilities
        prob_comparison = pd.DataFrame({
            'Sleep Disorder Class': class_names,
            'Baseline Probability': baseline_probs,
            'Improved Probability': improved_probs,
            'Absolute Change': improved_probs - baseline_probs,
            'Relative Change (%)': ((improved_probs - baseline_probs) / baseline_probs * 100)
        })

        # Display the table
        print("\nTable 1: Impact of AHA Simple 7 on Sleep Disorder Probabilities")
        print(tabulate(prob_comparison, headers='keys', tablefmt='grid', floatfmt='.4f'))

        # Create a table for the AHA Simple 7 habits and their physiological effects
        habit_effects = []
        for habit, changes in aha_habits:
            effect_str = ", ".join([f"{k}: {v:+.1f}" for k, v in changes.items()])
            habit_effects.append([habit, effect_str])

        habit_effects_table = pd.DataFrame(habit_effects, columns=['Habit', 'Physiological Effects'])

        print("\nTable 2: AHA Simple 7 Habits and Their Physiological Effects")
        print(tabulate(habit_effects_table, headers='keys', tablefmt='grid'))

        # Analyze each habit individually
        habit_results = {}

        for habit_name, changes in aha_habits:
            # Create a profile with just this habit's changes
            habit_profile = baseline_profile.copy()

            # Apply the changes for this habit
            for metric, change in changes.items():
                if metric in habit_profile:
                    # Apply bounds to keep values realistic
                    if metric in ['Quality of Sleep', 'Physical Activity Level']:
                        habit_profile[metric] = min(10, habit_profile[metric] + change)
                    elif metric == 'Sleep Duration':
                        habit_profile[metric] = min(9, habit_profile[metric] + change)
                    elif metric in ['Stress Level', 'BMI Category']:
                        habit_profile[metric] = max(1, habit_profile[metric] + change)
                    elif metric == 'Heart Rate':
                        habit_profile[metric] = max(55, habit_profile[metric] + change)
                    elif metric == 'Systolic':
                        habit_profile[metric] = max(100, habit_profile[metric] + change)
                    elif metric == 'Diastolic':
                        habit_profile[metric] = max(60, habit_profile[metric] + change)
                    elif metric == 'Daily Steps':
                        habit_profile[metric] = min(15000, habit_profile[metric] + change)
                    else:
                        habit_profile[metric] = habit_profile[metric] + change

            # Process through model pipeline
            habit_df = pd.DataFrame([habit_profile])
            habit_poly = poly.transform(habit_df)
            habit_scaled = scaler.transform(habit_poly)
            habit_tensor = torch.FloatTensor(habit_scaled)

            with torch.no_grad():
                habit_logits = model(habit_tensor)

            habit_probs = F.softmax(habit_logits, dim=1).numpy()[0]
            habit_results[habit_name] = habit_probs

        # Add "All Habits Combined" to the results
        habit_results["All Habits Combined"] = improved_probs

        # Create a stacked bar chart for each sleep disorder
        plt.figure(figsize=(14, 8))

        # For each habit, show its effect on each disorder
        habit_names = list(habit_results.keys())

        # Set up the plot
        fig, axes = plt.subplots(nrows=len(class_names), figsize=(12, 3*len(class_names)))

        for i, disorder in enumerate(class_names):
            ax = axes[i] if len(class_names) > 1 else axes

            # Get probabilities for this disorder across all habits
            disorder_probs = [habit_results[habit][i] for habit in habit_names]

            # Create horizontal bar chart
            bars = ax.barh(habit_names, disorder_probs, color='skyblue')
            ax.axvline(x=baseline_probs[i], color='red', linestyle='--',
                       label=f'Baseline: {baseline_probs[i]:.3f}')

            # Add value labels to the bars
            for bar in bars:
                width = bar.get_width()
                ax.text(width + 0.01, bar.get_y() + bar.get_height()/2,
                        f'{width:.3f}', ha='left', va='center')

            ax.set_title(f'Impact on Sleep Disorder Class {disorder}')
            ax.set_xlabel('Probability')
            ax.legend()

        plt.tight_layout()
        plt.show()

        # Create a table showing the impact of each habit on each sleep disorder
        habit_impact_data = []
        for i, disorder in enumerate(class_names):
            row = [f'Class {disorder}', f'{baseline_probs[i]:.4f}']
            for habit in habit_names:
                if habit != "All Habits Combined":
                    row.append(f'{habit_results[habit][i]:.4f}')
            row.append(f'{improved_probs[i]:.4f}')
            habit_impact_data.append(row)

        habit_impact_cols = ['Sleep Disorder', 'Baseline'] + [h for h in habit_names if h != "All Habits Combined"] + ['All Combined']
        habit_impact_table = pd.DataFrame(habit_impact_data, columns=habit_impact_cols)

        print("\nTable 3: Impact of Individual Habits on Sleep Disorder Probabilities")
        print(tabulate(habit_impact_table, headers='keys', tablefmt='grid'))


        # Calculate a neurodegeneration risk score based on sleep disorders
        # This is a hypothetical score based on research linking poor sleep to neurodegeneration
        def calculate_neuro_risk(probs, class_indices):
            # Weights based on research linking sleep disorders to neurodegeneration
            # Using numerical indices instead of string names
            # Assuming: 0 = None/Normal, 1 = Insomnia, 2 = Sleep Apnea, etc.
            # Adjust these indices based on your actual encoding
            weights = {}
            for i in class_indices:
                if i == 0:  # Assuming 0 is "None" or normal sleep
                    weights[i] = 0.05  # Healthy sleep is protective
                elif i == 1:  # Assuming 1 is "Insomnia"
                    weights[i] = 0.4   # Strong association with neurodegeneration
                elif i == 2:  # Assuming 2 is "Sleep Apnea"
                    weights[i] = 0.35  # Strong association with vascular dementia
                else:
                    weights[i] = 0.2   # Other disorders have moderate associations

            # Calculate weighted risk score
            risk_score = sum(probs[i] * weights[i] for i in range(len(probs)))
            return risk_score * 10  # Scale to 0-10 for easier interpretation

        # Calculate neurodegeneration risk for baseline and all habits
        baseline_risk = calculate_neuro_risk(baseline_probs, class_names)
        habit_risks = {habit: calculate_neuro_risk(probs, class_names)
                      for habit, probs in habit_results.items()}

        # Create a bar chart of neurodegeneration risk by habit
        plt.figure(figsize=(12, 6))

        # Sort habits by risk score (ascending)
        sorted_habits = sorted(habit_risks.items(), key=lambda x: x[1])
        habit_names = [h[0] for h in sorted_habits]
        risk_scores = [h[1] for h in sorted_habits]

        # Create the bar chart
        bars = plt.barh(habit_names, risk_scores, color='lightgreen')
        plt.axvline(x=baseline_risk, color='red', linestyle='--',
                   label=f'Baseline Risk: {baseline_risk:.2f}')

        # Add value labels
        for bar in bars:
            width = bar.get_width()
            plt.text(width + 0.1, bar.get_y() + bar.get_height()/2,
                    f'{width:.2f}', ha='left', va='center')

        plt.xlabel('Neurodegeneration Risk Score (0-10)')
        plt.title('Estimated Impact of AHA Simple 7 on Neurodegeneration Risk')
        plt.legend()
        plt.tight_layout()
        plt.show()

        # Create a table for neurodegeneration risk by habit
        risk_data = []
        for habit, risk in sorted_habits:
            risk_reduction = baseline_risk - risk
            risk_reduction_pct = (risk_reduction / baseline_risk) * 100
            risk_data.append([habit, f"{risk:.2f}", f"{risk_reduction:.2f}", f"{risk_reduction_pct:.1f}%"])

        risk_table = pd.DataFrame(risk_data, columns=['Habit', 'Risk Score (0-10)', 'Risk Reduction', 'Reduction (%)'])

        print("\nTable 4: Neurodegeneration Risk Reduction by Habit")
        print(tabulate(risk_table, headers='keys', tablefmt='grid'))


        # Create a heatmap showing the relationship between blood pressure and physical activity
        if 'Systolic' in baseline_profile and 'Physical Activity Level' in baseline_profile:
            # Define the ranges to explore
            bp_reductions = np.linspace(0, -15, 6)  # 0 to -15 mmHg reduction
            activity_increases = np.linspace(0, 3, 6)  # 0 to +3 activity level increase

            # Create a grid to store results
            risk_grid = np.zeros((len(activity_increases), len(bp_reductions)))

            # Calculate risk for each combination
            for i, activity_change in enumerate(activity_increases):
                for j, bp_change in enumerate(bp_reductions):
                    # Create profile with these specific changes
                    combo_profile = baseline_profile.copy()
                    combo_profile['Systolic'] = max(100, baseline_profile['Systolic'] + bp_change)
                    combo_profile['Diastolic'] = max(60, baseline_profile['Diastolic'] + bp_change * 0.6)
                    combo_profile['Physical Activity Level'] = min(10, baseline_profile['Physical Activity Level'] + activity_change)

                    # Process through model pipeline
                    combo_df = pd.DataFrame([combo_profile])
                    combo_poly = poly.transform(combo_df)
                    combo_scaled = scaler.transform(combo_poly)
                    combo_tensor = torch.FloatTensor(combo_scaled)

                    with torch.no_grad():
                        combo_logits = model(combo_tensor)

                    combo_probs = F.softmax(combo_logits, dim=1).numpy()[0]
                    risk_grid[i, j] = calculate_neuro_risk(combo_probs, class_names)

            # Create the heatmap
            plt.figure(figsize=(10, 8))
            sns.heatmap(risk_grid,
                        xticklabels=[f"{x:.1f}" for x in bp_reductions],
                        yticklabels=[f"{x:.1f}" for x in activity_increases],
                        cmap="YlGnBu_r", annot=True, fmt=".2f")
            plt.xlabel('Systolic Blood Pressure Reduction (mmHg)')
            plt.ylabel('Physical Activity Level Increase')
            plt.title('Combined Effect on Neurodegeneration Risk Score')
            plt.tight_layout()
            plt.show()

            # Extract key points from the heatmap
            min_risk_idx = np.unravel_index(np.argmin(risk_grid), risk_grid.shape)
            max_improvement_bp = bp_reductions[min_risk_idx[1]]
            max_improvement_activity = activity_increases[min_risk_idx[0]]
            min_risk_value = risk_grid[min_risk_idx]

            # Create a table with key findings
            combo_data = [
                ["Baseline Risk", f"{baseline_risk:.2f}", "N/A", "N/A"],
                ["No Changes", f"{risk_grid[0,0]:.2f}", f"{baseline_risk - risk_grid[0,0]:.2f}", f"{((baseline_risk - risk_grid[0,0])/baseline_risk)*100:.1f}%"],
                ["Max BP Reduction Only", f"{risk_grid[0,-1]:.2f}", f"{baseline_risk - risk_grid[0,-1]:.2f}", f"{((baseline_risk - risk_grid[0,-1])/baseline_risk)*100:.1f}%"],
                ["Max Activity Increase Only", f"{risk_grid[-1,0]:.2f}", f"{baseline_risk - risk_grid[-1,0]:.2f}", f"{((baseline_risk - risk_grid[-1,0])/baseline_risk)*100:.1f}%"],
                ["Optimal Combination", f"{min_risk_value:.2f}", f"{baseline_risk - min_risk_value:.2f}", f"{((baseline_risk - min_risk_value)/baseline_risk)*100:.1f}%"]
            ]

            combo_table = pd.DataFrame(combo_data, columns=['Scenario', 'Risk Score', 'Risk Reduction', 'Reduction (%)'])

            print("\nTable 5: Combined Effects of Blood Pressure and Physical Activity")
            print(tabulate(combo_table, headers='keys', tablefmt='grid'))


        # Print key findings and recommendations
        print("\n=== KEY FINDINGS ===")
        print(f"Baseline neurodegeneration risk score: {baseline_risk:.2f}")
        print(f"Risk with all AHA Simple 7 habits: {habit_risks['All Habits Combined']:.2f}")
        print(f"Risk reduction: {((baseline_risk - habit_risks['All Habits Combined']) / baseline_risk * 100):.1f}%")

        # Find the most effective individual habit
        individual_habits = {h: s for h, s in habit_risks.items() if h != "All Habits Combined"}
        most_effective = min(individual_habits.items(), key=lambda x: x[1])
        print(f"\nMost effective individual habit: {most_effective[0]}")
        print(f"Risk reduction from this habit alone: {((baseline_risk - most_effective[1]) / baseline_risk * 100):.1f}%")

        print("\n=== RECOMMENDATIONS BASED ON AHA'S SIMPLE 7 ===")
        print("1. Blood Pressure Management: Maintaining healthy blood pressure shows the strongest individual effect on reducing sleep disorders associated with neurodegeneration risk.")
        print("2. Regular Exercise: Physical activity improvements correlate with significant reductions in sleep disorder probabilities and estimated neurodegeneration risk.")
        print("3. Healthy Diet: Dietary improvements that reduce BMI show substantial benefits for sleep quality and potential neurodegeneration risk.")
        print("4. Combined Approach: The model suggests that the combination of all AHA Simple 7 habits provides significantly greater benefits than any single habit alone.")
        print("5. Sleep Quality: Improvements in sleep quality appear to be both outcomes of better cardiovascular health and potential mediators of reduced neurodegeneration risk.")

        return {
            'baseline_probs': baseline_probs,
            'improved_probs': improved_probs,
            'habit_results': habit_results,
            'baseline_risk': baseline_risk,
            'habit_risks': habit_risks,
            'risk_grid': risk_grid if 'risk_grid' in locals() else None
        }

    except Exception as e:
        print(f"Error during analysis: {str(e)}")
        import traceback
        traceback.print_exc()
        return None

results = analyze_heart_health_neuro_impact(
    model=final_model,
    data=data,
    poly=poly,
    scaler=scaler,
    le_target=le_target
)
```
